<h3>What a Matrix Is</h3>
<p>I have spent two weeks working with vectors. A vector is a single ordered list of numbers, each one describing a different attribute of one thing: one employee's tenure, performance score, absence days, and engagement rating, all collected into a single object. That is powerful, but it only describes one thing at a time.</p>
<p>A <span class="term">matrix</span> is what happens when I stack multiple vectors together. If a vector is one employee's profile, a matrix is the entire team's profiles collected into a single structure, with every employee in their own row and every attribute in its own column. At this stage, a matrix is best understood as a way of organising many vectors together in one structure.</p>
<p>Formally, a matrix is a rectangular array of numbers arranged in <span class="term">rows</span> and <span class="term">columns</span>. The <span class="term">shape</span> of a matrix is written as m × n, where m is the number of rows and n is the number of columns. A 4 × 4 matrix has four rows and four columns. A 500 × 12 matrix has 500 rows and 12 columns. The row count always comes first.</p>
<p>Individual numbers inside a matrix are called <span class="term">elements</span>. The notation for locating an element uses the row index first and the column index second, so the element in row 2, column 3 is written A<sub>23</sub>. Row first, column second, always.</p>
<p>A vector is a special case of a matrix. A <span class="term">row vector</span> is a 1 × n matrix (one row, many columns). A <span class="term">column vector</span> is an m × 1 matrix (many rows, one column). Everything learned about vectors applies directly. A matrix generalises the idea to hold many vectors at once.</p>
<div class="inner-formal">
<div class="inner-formal-label">Formal Notation</div>
<p>A matrix is written as a bold capital letter such as <strong>A</strong>. The element in row i and column j is written A<sub>ij</sub>, so A<sub>23</sub> means the value in row 2, column 3. The shape of a matrix with m rows and n columns is written m × n, read as "m by n". A single row extracted from <strong>A</strong> is a row vector. A single column extracted from <strong>A</strong> is a column vector.</p>
<p>In NumPy, the same position is accessed with zero-based indexing, so the element at row 2, column 3 in mathematical notation is written <code class="inline">A[1, 2]</code> in code. The maths counts from one. Python counts from zero.</p>
</div>
<p>In NumPy, I construct a matrix using <code class="inline">np.array</code> with a list of lists. Each inner list is one row. The shape is available via <code class="inline">.shape</code>, which returns (rows, columns). I access a single element with <code class="inline">A[i, j]</code>, a full row with <code class="inline">A[i]</code>, and a full column with <code class="inline">A[:, j]</code>.</p>
</div>
<div class="panel">
<div class="accent-strip accent-amber"></div>
<div class="sec-row">
<div class="sec-label amber">AI Connection</div>
<div class="sec-rule"></div>
</div>
<p>In machine learning, the training data is stored as a <span class="term">feature matrix</span> <strong>X</strong>, where each row is one example and each column is one feature. In neural networks, the learned connections between layers are stored in <span class="term">weight matrices</span>. Matrices are not just a way to organise data. They are one of the main structures through which models learn and make predictions.</p>
</div>
<div class="panel">
<div class="accent-strip accent-amber"></div>
<div class="sec-row">
<div class="sec-label amber">Workforce Analytics</div>
<div class="sec-rule"></div>
</div>
<p>Every structured workforce dataset I work with is a matrix. Employees occupy the rows and attributes occupy the columns. Tenure, performance rating, absence days, and engagement score each become a column, and each employee becomes a row. The shape of the matrix tells me something useful before I have looked at a single number. A 500 × 12 matrix means I have 500 employees each described by 12 metrics.</p>
<p>A different kind of matrix appears in headcount reporting: departments as rows, months as columns, and headcount figures as elements. These are aggregate matrices rather than employee-level ones, but the structure is the same. In both cases, knowing the shape, navigating to a row, and extracting a column are the fundamental operations.</p>
<p>When I run any analytical model in people data work, from a simple correlation to a regression to a clustering algorithm, the first thing that model receives is a matrix. Being able to construct one, inspect it, navigate its rows and columns, and understand what its shape is telling me is not a preliminary step. It is the work itself.</p>
</div>
<div class="panel">
<div class="accent-strip accent-green"></div>
<div class="sec-row">
<div class="sec-label green">Worked Example</div>
<div class="sec-rule"></div>
</div>
<p>Four employees, four attributes: tenure in years, performance score (1 to 10), absence days in the past year, and engagement score (1 to 10). The code below builds this as a NumPy matrix, inspects its shape, accesses individual elements, and extracts full rows and columns.</p>
<p>The output covers five navigation operations in sequence: shape, element access, row extraction, column extraction, and a simple column aggregate. These are the operations I will reach for constantly when working with any structured dataset.</p>
</div>
</div>

```python
import numpy as np

# ── Workforce matrix ───────────────────────────────────────────────────────────
# Rows    : employees A, B, C, D
# Columns : tenure (yrs) | performance (1-10) | absence (days) | engagement (1-10)

workforce = np.array([
    [3,  7,  4,  8],
    [7,  9,  2,  9],
    [1,  5, 11,  5],
    [5,  8,  6,  7],
], dtype = float)

# ── Shape ──────────────────────────────────────────────────────────────────────
print(f"Shape:    {workforce.shape}")
print(f"Rows:     {workforce.shape[0]}  (employees)")
print(f"Columns:  {workforce.shape[1]}  (attributes)")

# ── Element access ─────────────────────────────────────────────────────────────
print(f"\nEmployee A, absence days [0, 2]:  {workforce[0, 2]:.0f}")
print(f"Employee C, performance  [2, 1]:  {workforce[2, 1]:.0f}")

# ── Row access: full employee profiles ────────────────────────────────────────
print(f"\nEmployee A profile: {workforce[0]}")
print(f"Employee C profile: {workforce[2]}")

# ── Column access: one attribute across all employees ─────────────────────────
print(f"\nPerformance scores (all employees): {workforce[:, 1]}")
print(f"Absence days       (all employees): {workforce[:, 2]}")

# ── Column summary ─────────────────────────────────────────────────────────────
print(f"\nMean performance:   {np.mean(workforce[:, 1]):.2f}")
print(f"Total absence days: {np.sum(workforce[:, 2]):.0f}")
```
