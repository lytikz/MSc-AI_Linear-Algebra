<div class="sec-label green">Concept</div>
<div class="sec-rule"></div>
</div>
<h3>What Happens When Transformations Are Chained</h3>
<p>In Lesson 2, a single transformation matrix was applied to a set of employee vectors. The result was a new set of vectors in a new space. That is the foundation. In this lesson the question becomes: what happens when I apply one transformation and then apply another? What is the combined effect, and is there a single matrix that captures it?</p>
<p>The answer is yes, and it falls directly out of the associativity of matrix multiplication. If I apply transformation A to a dataset X to get XA, and then apply transformation B to that result to get (XA)B, the result is the same as applying the single matrix AB to the original data: X(AB). The <span class="term">combined transformation</span> is the product of the individual transformation matrices.</p>
<p>This is not a convenient shorthand. It is a fundamental property of linear transformations. Any sequence of linear transformations can be collapsed into a single matrix. That matrix encodes the entire pipeline in one operation.</p>
<p>There is one critical thing to get right: the order. Applying A then B produces the combined matrix AB. Applying B then A produces BA. These are generally different matrices and generally produce different results. The order of multiplication encodes the order of application, and reversing it changes the outcome.</p>
<p>A concrete way to see this is with two transformations that do not commute. A rotation followed by a scaling produces a different result than the same scaling followed by the same rotation. The combined matrix captures that distinction exactly. If I multiply the wrong way round, I am computing the wrong pipeline.</p>
<p>This chaining property is also why depth in a neural network only adds expressive power because of the non-linear activation functions between layers. Without them, any chain of linear transformations would collapse to a single matrix multiplication regardless of depth, because any product of linear transformations is itself a linear transformation. Stacking layers without activations would accomplish nothing that a single layer could not.</p>
<div class="inner-formal">
<div class="inner-formal-label">Formal Notation</div>
<p>Given matrices A of shape (k, m) and B of shape (m, p), the chained transformation applied to X of shape (n, k) is X(AB) = (XA)B. <span class="term">Associativity</span> guarantees these are equal. The combined matrix C = AB has shape (k, p) and represents the full pipeline in one object. Computing C first and then applying it once is more efficient than applying A and B in sequence, particularly when the same pipeline is applied to many datasets.</p>
</div>
</div>
<div class="panel">
<div class="accent-strip accent-amber"></div>
<div class="sec-row">
<div class="sec-label amber">AI Connection</div>
<div class="sec-rule"></div>
</div>
<p>The chaining property has a direct consequence for how neural networks are analysed. If there were no non-linearities, the entire forward pass of a deep network would reduce to a single matrix multiplication regardless of how many layers it had. Adding layers without activation functions adds no representational power at all, because any product of linear transformations is itself a linear transformation.</p>
<p>This is why activation functions are not optional. They are the mechanism that breaks the chaining collapse and allows deeper networks to represent more complex functions than shallower ones. Understanding chaining and its limits is understanding why network depth matters and what it actually buys.</p>
<p>The same idea appears in dimensionality reduction. PCA applies a projection matrix to a dataset. If the data has already been scaled by a linear transformation, the two steps can be combined into a single matrix product. In practice they are often applied separately for interpretability, but the mathematics is one chain.</p>
</div>
<div class="panel">
<div class="accent-strip accent-amber"></div>
<div class="sec-row">
<div class="sec-label amber">Workforce Analytics</div>
<div class="sec-rule"></div>
</div>
<p>A realistic people analytics workflow often combines a preprocessing shift with linear transformation steps. After centring the data by subtracting column means, I might apply a scaling matrix and then a projection matrix. The linear part of that pipeline can be written as a chain of matrix multiplications, and the combined matrix encodes those transformation steps in order.</p>
<p>Understanding this as a chain has practical consequences. If the same pipeline needs to be applied to new data, for example a new cohort of employees being scored against a model built on historical data, the combined transformation matrix can be stored and reapplied directly without repeating each step.</p>
<p>Order matters here too. Scaling before projecting is not the same as projecting before scaling. Being explicit about the order of operations, and knowing that matrix multiplication order encodes that sequence, is part of working rigorously with multivariate data.</p>
</div>
<div class="panel">
<div class="accent-strip accent-green"></div>
<div class="sec-row">
<div class="sec-label green">Worked Example</div>
<div class="sec-rule"></div>
</div>
<p>Four employee vectors, each described by two attributes: standardised performance and standardised engagement. Two transformation matrices are defined: a scaling matrix S that stretches the performance axis by a factor of 2, and a rotation matrix R that rotates all vectors by 30 degrees anticlockwise using the row vector convention.</p>
<p>The code applies them in both orders, scale then rotate and rotate then scale, and shows that the results differ. It then confirms that each chained result matches the corresponding single combined matrix. Four side by side plots show the original vectors alongside three transformed versions so the difference between the two orderings is visible directly.</p>
</div>
</div>

```python
import numpy             as np
import matplotlib.pyplot as plt

# ── Colour constants ───────────────────────────────────────────────────────────
BG    = '#0a0a0c'
TEXT  = '#e8e4de'
SUB   = '#b5b0a8'
MUTED = '#6a6560'
GRID  = '#2a2a28'
AXIS  = '#3a3a38'
GRN   = '#c4f042'
AMB   = '#f0a830'

# ── Employee vectors (4 employees x 2 attributes) ─────────────────────────────
# Columns : standardised performance | standardised engagement

employees = np.array([
    [ 1.0,  0.5],
    [-0.5,  1.2],
    [ 0.8, -0.9],
    [-1.1, -0.4],
], dtype = float)

names = ['A', 'B', 'C', 'D']

# ── Transformation S: scale performance axis by 2 ─────────────────────────────
S = np.array([
    [2.0, 0.0],
    [0.0, 1.0],
], dtype = float)

# ── Transformation R: 30-degree anticlockwise rotation (row vector convention) ─
theta = np.radians(30)
R     = np.array([
    [ np.cos(theta), np.sin(theta)],
    [-np.sin(theta), np.cos(theta)],
], dtype = float)

print(f"S (scaling) shape:   {S.shape}")
print(f"R (rotation) shape:  {R.shape}")

# ── Apply in both orders ───────────────────────────────────────────────────────
scale_then_rotate = employees @ S @ R
rotate_then_scale = employees @ R @ S

# ── Confirm chaining matches combined matrix ───────────────────────────────────
SR     = S @ R
RS     = R @ S
via_SR = employees @ SR
via_RS = employees @ RS

print(f"\nScale then rotate matches employees @ SR: {np.allclose(scale_then_rotate, via_SR)}")
print(f"Rotate then scale matches employees @ RS: {np.allclose(rotate_then_scale, via_RS)}")
print(f"SR == RS (order matters):                 {np.allclose(SR, RS)}")

# ── Print positions under both orderings ──────────────────────────────────────
print(f"\n{'Employee':<10} {'Scale->Rotate':>24} {'Rotate->Scale':>24}")
print(f"{'-' * 60}")
for i, name in enumerate(names):
    sr = f"[{scale_then_rotate[i,0]:+.3f}, {scale_then_rotate[i,1]:+.3f}]"
    rs = f"[{rotate_then_scale[i,0]:+.3f}, {rotate_then_scale[i,1]:+.3f}]"
    print(f"{name:<10} {sr:>24} {rs:>24}")

# ── Four side by side plots ────────────────────────────────────────────────────
fig, axes = plt.subplots(
    1, 4,
    figsize   = (18, 5),
    facecolor = BG
)

datasets = [employees, employees @ S, scale_then_rotate, rotate_then_scale]
titles   = [
    'Original',
    'After scaling',
    'Scale then rotate',
    'Rotate then scale'
]
colours  = [GRN, AMB, GRN, AMB]

for ax, data, title, col in zip(axes, datasets, titles, colours):
    ax.set_facecolor(BG)
    ax.set_aspect('equal', adjustable = 'box')
    ax.axhline(0, color = AXIS, linewidth = 0.8)
    ax.axvline(0, color = AXIS, linewidth = 0.8)
    ax.grid(
        True,
        color     = GRID,
        linewidth = 0.5,
        linestyle = '--'
    )

    for i, name in enumerate(names):
        ax.quiver(
            0, 0, data[i, 0], data[i, 1],
            angles      = 'xy',
            scale_units = 'xy',
            scale       = 1,
            color       = col,
            width       = 0.012,
            alpha       = 0.85
        )
        ax.annotate(
            name,
            xy     = (data[i, 0], data[i, 1]),
            xytext = (data[i, 0] + 0.06, data[i, 1] + 0.06),
            color  = TEXT,
            fontsize = 9
        )

    ax.set_title(
        title,
        color    = TEXT,
        fontsize = 10,
        pad      = 12
    )
    ax.set_xlabel("Performance", color = SUB, fontsize = 8)
    ax.set_ylabel("Engagement",  color = SUB, fontsize = 8)
    ax.tick_params(colors = MUTED)
    for spine in ax.spines.values():
        spine.set_edgecolor(AXIS)

plt.suptitle(
    "Chaining transformations: scale then rotate vs rotate then scale",
    color    = TEXT,
    fontsize = 11,
    y        = 1.02
)
plt.tight_layout()
plt.show()
```
