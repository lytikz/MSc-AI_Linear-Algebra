<h3>Same Structure, Richer Question</h3>
<p>In Lesson 1 I asked what a vector is. In Lesson 2 I asked what it means to compare two vectors using the dot product. In this lesson the question becomes: what can I actually conclude from a vector comparison in a real analytical context, and when does my comparison tool need to change?</p>
<p>The context shifts to absence patterns. Each department's monthly absence rate across a 12-month period is represented as a <span class="term">12-dimensional vector</span>, one component per month. Two departments can now be compared not just by how much absence they have in total but by the shape of their pattern across the year.</p>
<p>A department with absence concentrated in winter tells a very different story from one with absence spread evenly year-round, even if their total volumes are similar. That distinction matters analytically. It points towards different causes and different interventions.</p>
<p>This is where the dot product runs into a problem. Two departments with very different total absence volumes will produce a misleading dot product score even if their seasonal patterns are almost identical. The magnitude difference dominates the result and obscures the pattern comparison.</p>
<p>A better comparison tool here is <span class="term">cosine similarity</span>: divide the dot product by both magnitudes. This reduces the effect of overall vector length, so I can compare pattern direction rather than raw volume alone. Two vectors pointing in the same direction score 1.0 regardless of how long they are.</p>
<div class="inner-card">
<div class="inner-card-label">Perspective</div>
<p>This lesson is the first time the vectors are genuinely high-dimensional. The move from 2D to 12D means there is no direct arrow plot to fall back on. The line chart replaces it — not as an approximation but as the honest visualisation. This is how most real data vectors are understood: through the patterns they trace, not through geometry I can draw on a page.</p>
</div>
<div class="inner-note">
<div class="inner-note-label">Professional Note</div>
<p>Cosine similarity is a better comparison tool here, not a perfect one. It reduces the effect of magnitude but does not resolve every scaling issue. If the underlying features were on very different scales, or if the data had not already been expressed as rates, further standardisation would be needed. That is a later topic.</p>
</div>
<div class="inner-formal">
<div class="inner-formal-label">Formal Notation</div>
<p>For two vectors <strong>a</strong> and <strong>b</strong>, cosine similarity is defined as:</p>
<p><strong>cos(a, b)</strong> = (<strong>a</strong> &middot; <strong>b</strong>) / (|<strong>a</strong>| &times; |<strong>b</strong>|)</p>
<p>In NumPy: <code class="inline">np.dot(a, b) / (np.linalg.norm(a) * np.linalg.norm(b))</code></p>
<p>Range: 1.0 = identical direction, 0.0 = unrelated, -1.0 = opposite direction.</p>
</div>

<h3>Semantic Search and Embedding Comparison</h3>
<p>When a language model compares two pieces of text, it converts each into a high-dimensional <span class="term">embedding vector</span> and computes the cosine similarity between them. A short query and a long document that are semantically related will have embedding vectors pointing in similar directions even though one is far longer than the other. Cosine similarity handles that length difference automatically.</p>
<p>This is the direct application of what I am learning here, just in hundreds or thousands of dimensions rather than twelve. The mathematics is identical. The scale is not.</p>
</div>

<h3>Absence Pattern Analysis Across Departments</h3>
<p>Knowing that two departments have similar absence rates in absolute terms is useful but limited. Knowing that they share a similar pattern across the year is more analytically powerful, because it suggests shared causes rather than department-specific ones. A winter peak appearing across multiple departments might point to seasonal illness or a structural rota issue rather than anything specific to one team.</p>
<p>Cosine similarity makes that pattern comparison possible across many departments without being driven mainly by differences in overall absence volume. The shape of the pattern is what matters, and cosine similarity is designed to surface exactly that.</p>
<div class="inner-note">
<div class="inner-note-label">Professional Note</div>
<p>In practice, this kind of pattern comparison underlies clustering approaches in people analytics, where departments or individuals are grouped not by their absolute values but by the shape of their behaviour over time. The technique scales naturally from two departments to two hundred.</p>
</div>
</div>

</div>

```python
import numpy             as np
import matplotlib.pyplot as plt

# ── Colour constants ───────────────────────────────────────────────────────────
BG    = '#0a0a0c'
TEXT  = '#e8e4de'
SUB   = '#b5b0a8'
MUTED = '#6a6560'
GRID  = '#2a2a28'
AXIS  = '#3a3a38'
GRN   = '#c4f042'
AMB   = '#f0a830'
RED   = '#e8453c'

# ── Helper function ────────────────────────────────────────────────────────────
def cosine_similarity(a, b):
    return np.dot(a, b) / (np.linalg.norm(a) * np.linalg.norm(b))

# ── Absence pattern vectors ────────────────────────────────────────────────────
# Monthly absence rates (%) across 12 months for two departments.
# Index 0 = January, index 11 = December.

months = ['Jan','Feb','Mar','Apr','May','Jun',
          'Jul','Aug','Sep','Oct','Nov','Dec']

dept_a = np.array([2.1, 1.8, 1.5, 1.2, 1.0, 0.9, 0.8, 0.9, 1.1, 1.6, 2.3, 2.8], dtype = float)
dept_b = np.array([1.1, 1.0, 1.2, 1.3, 1.4, 1.3, 1.2, 1.3, 1.4, 1.2, 1.1, 1.0], dtype = float)

# ── Magnitudes ────────────────────────────────────────────────────────────────
mag_a = np.linalg.norm(dept_a)
mag_b = np.linalg.norm(dept_b)

print("Absence pattern vectors")
print(f"  Department A: {dept_a}")
print(f"  Department B: {dept_b}")

print("\nMagnitudes (overall absence intensity)")
print(f"  Department A: {mag_a:.4f}")
print(f"  Department B: {mag_b:.4f}")

# ── Dot product ───────────────────────────────────────────────────────────────
dot_ab = np.dot(dept_a, dept_b)

print("\nDot product")
print(f"  A · B:        {dot_ab:.4f}")
print(f"  (scale-sensitive — magnitudes differ, so this alone is not enough)")

# ── Cosine similarity ─────────────────────────────────────────────────────────
cos_ab = cosine_similarity(dept_a, dept_b)

print("\nCosine similarity")
print(f"  cos(A, B):    {cos_ab:.4f}")

# ── Visualisation ─────────────────────────────────────────────────────────────
x = np.arange(len(months))

fig, ax = plt.subplots(figsize = (9, 5))
fig.patch.set_facecolor(BG)
ax.set_facecolor(BG)

ax.plot(
    x, dept_a,
    color      = GRN,
    linewidth  = 1.8,
    marker     = 'o',
    markersize = 4,
    label      = 'Department A'
)

ax.plot(
    x, dept_b,
    color      = AMB,
    linewidth  = 1.8,
    marker     = 'o',
    markersize = 4,
    label      = 'Department B'
)

ax.set_xticks(x)
ax.set_xticklabels(months, color = MUTED, fontsize = 8)
ax.set_ylabel("Absence Rate (%)", color = SUB,  fontsize = 9)
ax.tick_params(colors = MUTED)
ax.grid(True, color = GRID, linewidth = 0.5, linestyle = '--')

for spine in ax.spines.values():
    spine.set_edgecolor(AXIS)

ax.legend(
    facecolor  = '#111110',
    edgecolor  = AXIS,
    labelcolor = SUB,
    fontsize   = 9
)

ax.set_title(
    "Monthly Absence Patterns — Department A vs Department B",
    color    = TEXT,
    fontsize = 11,
    pad      = 14
)

plt.tight_layout()
plt.show()
```
