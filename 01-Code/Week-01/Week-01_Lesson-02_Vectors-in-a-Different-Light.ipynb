<h3>The Same Structure, a Different Question</h3>
<p>In Lesson 1, I looked at a single employee profile vector and asked: what is this? I understood what a vector is, what its components represent, and what its magnitude measures.</p>
<p>In this lesson the question changes. Instead of asking what a vector is, I ask: what does it mean to compare two vectors?</p>
<p>The same vector structure can describe either what a person offers or what a role demands. The comparison between those two vectors is where <span class="term">analytical meaning</span> begins to emerge.</p>
<p>Consider a job role described by two qualities: analytical reasoning and stakeholder communication. I can represent that role as a vector — two numbers, in a fixed order, each one measuring a requirement on a known scale. A candidate can be described the same way. Now I have two vectors sitting in the same space, and a real question to ask: how close are they?</p>
<div class="inner-card">
<div class="inner-card-label">Perspective</div>
<p>In Lesson 1 the vector described a person. Here it describes a role. The mathematics does not change — but the analytical story does. Comparing two vectors is the foundation of every similarity and matching problem in data science, from recommendation engines to job fit models to nearest-neighbour classifiers.</p>
</div>
<p>As a first step towards comparing vectors, I introduce the <span class="term">dot product</span> as a simple way to think about similarity, framed plainly before any formal notation. The dot product multiplies matching components and adds them up. If two vectors point in a similar direction — if their features tend to move together — the result is large. If they point in very different directions, the result is smaller.</p>
<p>This is an early preview of a tool I will use properly in later weeks. For now, the goal is intuition: two vectors can be compared, and the dot product is one honest way to start.</p>
<div class="inner-note">
<div class="inner-note-label">Professional Note</div>
<p>Raw similarity scores can be misleading when one feature has a much larger numeric scale than the others. A candidate who scores highly on a 100-point scale will dominate the dot product against a candidate measured on a 10-point scale — even if the underlying match quality is the same. Scale awareness matters here, and I will address it properly in Lesson 3.</p>
</div>
<div class="inner-formal">
<div class="inner-formal-label">Formal Notation</div>
<p>For two vectors <strong>a</strong> = [a&#8321;, a&#8322;] and <strong>b</strong> = [b&#8321;, b&#8322;], the dot product is defined as:</p>
<p><strong>a &middot; b</strong> = a&#8321;b&#8321; + a&#8322;b&#8322;</p>
<p>In NumPy: <code class="inline">np.dot(a, b)</code></p>
</div>

<h3>How Language Models Measure Similarity</h3>
<p>Every word, sentence, or document processed by a modern language model is represented as a vector — a list of numbers called an <span class="term">embedding</span>. When a search engine or a chat model decides whether two pieces of text are related, it is comparing those vectors.</p>
<p>The dot product is one of the core tools for this comparison. A normalised version of it — <span class="term">cosine similarity</span> — is used throughout semantic search, retrieval-augmented generation, and recommendation systems. The geometry I am learning here is the same geometry those systems operate on, just in many more dimensions.</p>
</div>

<h3>Job Fit as a Vector Comparison Problem</h3>
<p>One of the classic problems in people analytics is job fit: given a candidate profile and a role requirements profile, how well does the candidate match? Structuring both as vectors and computing their similarity is a principled way to begin answering that question at scale.</p>
<p>In practice this is how many <span class="term">applicant tracking systems</span> and internal mobility tools work under the surface — comparing employee skill vectors against role requirement vectors to surface candidates, identify development gaps, or flag flight risks before they become vacancies.</p>
<p>The dot product gives a first similarity score. It is imperfect and scale-sensitive, but it is a real starting point — and understanding it properly is the foundation for understanding more sophisticated matching models later.</p>
</div>

</div>

```python
import numpy             as np
import matplotlib.pyplot as plt
import matplotlib        as mpl

# ── Colour constants ───────────────────────────────────────────────────────────
BG    = '#0a0a0c'
TEXT  = '#e8e4de'
SUB   = '#b5b0a8'
MUTED = '#6a6560'
GRID  = '#2a2a28'
AXIS  = '#3a3a38'
GRN   = '#c4f042'
AMB   = '#f0a830'
RED   = '#e8453c'

# ── Role and candidate vectors ─────────────────────────────────────────────────
# Two dimensions: analytical reasoning (x) and stakeholder communication (y).
# Scale: 0 to 10 for both dimensions.

role        = np.array([8, 6], dtype = float)   # what the role demands
candidate_a = np.array([7, 5], dtype = float)   # strong analytical, reasonable communication
candidate_b = np.array([3, 8], dtype = float)   # weaker analytical, strong communication

# ── Magnitudes ────────────────────────────────────────────────────────────────
mag_role        = np.linalg.norm(role)
mag_candidate_a = np.linalg.norm(candidate_a)
mag_candidate_b = np.linalg.norm(candidate_b)

print("Role and candidate profiles")
print(f"  Role vector:        {role}")
print(f"  Candidate A vector: {candidate_a}")
print(f"  Candidate B vector: {candidate_b}")

print("\nMagnitudes")
print(f"  Role magnitude:        {mag_role:.4f}")
print(f"  Candidate A magnitude: {mag_candidate_a:.4f}")
print(f"  Candidate B magnitude: {mag_candidate_b:.4f}")

# ── Dot products ──────────────────────────────────────────────────────────────
dot_a = np.dot(role, candidate_a)
dot_b = np.dot(role, candidate_b)

print("\nDot products (role vs each candidate)")
print(f"  Role · Candidate A: {dot_a}")
print(f"  Role · Candidate B: {dot_b}")
print(f"  ({'Candidate A' if dot_a >= dot_b else 'Candidate B'} has the higher raw dot product score)")

# ── Visualisation ─────────────────────────────────────────────────────────────
fig, ax = plt.subplots(figsize = (7, 6))
fig.patch.set_facecolor(BG)
ax.set_facecolor(BG)

ax.set_xlim(-0.5, 10.5)
ax.set_ylim(-0.5, 10.5)
ax.set_aspect('equal', adjustable = 'box')

ax.axhline(0, color = AXIS, linewidth = 0.8)
ax.axvline(0, color = AXIS, linewidth = 0.8)
ax.grid(True, color = GRID, linewidth = 0.5, linestyle = '--')

for vec, col, label in [
    (role,        GRN, 'Role'),
    (candidate_a, AMB, 'Candidate A'),
    (candidate_b, RED, 'Candidate B'),
]:
    ax.quiver(
        0, 0, vec[0], vec[1],
        angles      = 'xy',
        scale_units = 'xy',
        scale       = 1,
        color       = col,
        width       = 0.012,
        zorder      = 3
    )
    ax.text(
        vec[0] + 0.2, vec[1] + 0.2,
        label,
        color    = col,
        fontsize = 9
    )

ax.set_xlabel("Analytical Reasoning",      color = SUB, fontsize = 9)
ax.set_ylabel("Stakeholder Communication", color = SUB, fontsize = 9)
ax.tick_params(colors = MUTED)

for spine in ax.spines.values():
    spine.set_edgecolor(AXIS)

ax.set_title(
    "Role vs Candidate Vectors",
    color    = TEXT,
    fontsize = 11,
    pad      = 14
)

plt.tight_layout()
plt.show()
```
