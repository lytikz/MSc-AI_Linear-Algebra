<h3>Removing accidental scale, then adding deliberate weight</h3>
<p>Across Lessons 1 and 2, the same problem has been lurking beneath the surface. The employee profile vector mixes dimensions that sit on completely different scales: age in the range of 20 to 60, tenure in the range of 0 to 30, engagement scored from 0 to 100, and performance rated from 1 to 5. When I compute a magnitude or a Euclidean distance, the engagement dimension dominates simply because its numbers are largest. That is not a meaningful result. It is a measurement artefact.</p>
<p>There are two related but distinct tools for addressing this: <span class="term">normalisation</span> and <span class="term">weighting</span>. They solve different problems and it is worth being precise about the difference.</p>
<p><span class="term">Min-max normalisation</span> puts every dimension on a comparable scale so that no single dimension dominates by virtue of its units. It rescales each dimension to the range 0 to 1. If I know the minimum and maximum plausible values for each feature, I subtract the minimum and divide by the range. Every dimension now sits between 0 and 1, and Euclidean distance becomes a more honest comparison.</p>
<p>A second common approach is <span class="term">unit vector normalisation</span>, also called normalising to unit length. I divide a vector by its own magnitude, producing a vector that points in the same direction but has magnitude exactly 1. This is not about making features comparable in scale &mdash; it is about separating direction from size. When I compute cosine similarity, I am implicitly working with unit vectors. Making that explicit is sometimes useful.</p>
<p><span class="term">Weighting</span> is a different operation. It uses element-wise multiplication to amplify or reduce the influence of specific dimensions according to their analytical importance. If performance rating matters twice as much as age for a particular scoring task, I can multiply the performance dimension by 2 before computing distance. That is a deliberate analytical choice, not a correction for scale. Weighting assumes the dimensions are already on comparable scales, or that I am consciously building the scale difference into my scoring model.</p>
<p>These two operations can be combined. I normalise first to remove accidental scale dominance, then weight deliberately to reflect analytical priorities. Summing the weighted normalised values gives a single <span class="term">composite score</span> &mdash; a defensible, scale-aware summary of a profile.</p>
<div class="inner-formal">
<div class="inner-formal-label">Formal Notation</div>
<p>For a single dimension with value x, minimum m, and maximum M:</p>
<p>Min-max normalisation: &nbsp;&nbsp; x&#x0302; = (x &#8722; m) / (M &#8722; m)</p>
<p>For a vector <strong>v</strong> with magnitude &#8214;<strong>v</strong>&#8214;:</p>
<p>Unit vector normalisation: &nbsp;&nbsp; <strong>v&#x0302;</strong> = <strong>v</strong> / &#8214;<strong>v</strong>&#8214;</p>
<p>For a weight vector <strong>w</strong> applied element-wise to a normalised profile vector <strong>v&#x0302;</strong>:</p>
<p>Element-wise weighted vector: &nbsp;&nbsp; <strong>w</strong> &#x2299; <strong>v&#x0302;</strong> = [w&#8321;v&#x0302;&#8321;, w&#8322;v&#x0302;&#8322;, ..., w&#8345;v&#x0302;&#8345;]</p>
<p>Composite score: &nbsp;&nbsp; s = w&#8321;v&#x0302;&#8321; + w&#8322;v&#x0302;&#8322; + ... + w&#8345;v&#x0302;&#8345;</p>

<h3>Why scaling decisions define model performance</h3>
<p>Normalisation is not optional in most machine learning pipelines &mdash; it is a prerequisite. Neural networks train significantly faster and more stably when input features are on comparable scales, because large-valued features otherwise dominate gradient updates during backpropagation. <span class="term">Batch normalisation</span>, a technique used inside deep networks, applies a similar principle layer by layer during training. In natural language processing, unit vector normalisation of word or sentence embeddings is standard practice before computing cosine similarity, because it ensures the similarity measure reflects meaning rather than vector magnitude. Feature scaling decisions &mdash; whether to normalise, standardise, or leave raw &mdash; are one of the most consequential choices in building a machine learning model, and getting them wrong silently degrades performance.</p>
</div>

<div class="panel">
<div class="accent-strip accent-amber"></div>
<div class="sec-row">
<div class="sec-label amber">Workforce Analytics Connection</div>
<div class="sec-rule"></div>
</div>
<h3>Building defensible scoring models</h3>
<p>In workforce analytics, weighting and normalisation together form the backbone of any composite scoring model. An engagement index, a flight risk score, or a performance potential rating all involve taking multiple dimensions, putting them on a common scale, and then applying weights that reflect the analytical or business judgement of what matters most. The weights are where the analytical assumptions live &mdash; and they should be explicit and defensible rather than implicit in the raw scale of the data. A dimension that dominates a score because its numbers happen to be large is not a weighted model. It is an unexamined one. Understanding this distinction is what separates a people analytics practitioner who builds models from one who merely runs them.</p>
</div>

</div>


```python
import numpy             as np
import matplotlib.pyplot as plt

# ── Colour constants ──────────────────────────────────────────────────────────
BG    = '#0a0a0c'
TEXT  = '#e8e4de'
SUB   = '#b5b0a8'
MUTED = '#6a6560'
GRID  = '#2a2a28'
AXIS  = '#3a3a38'
GRN   = '#c4f042'
AMB   = '#f0a830'
RED   = '#e8453c'

# ── Two employee profiles ─────────────────────────────────────────────────────
# Four dimensions: age, tenure (years), engagement score (0-100), performance (1-5)
employee_a = np.array([34,  5, 72, 4], dtype = float)
employee_b = np.array([29, 11, 58, 3], dtype = float)
labels     = ['Age', 'Tenure', 'Engagement', 'Performance']

# ── Min-max bounds per dimension ──────────────────────────────────────────────
# Set based on plausible organisational ranges, not the sample itself
dim_min = np.array([20,  0,   0, 1], dtype = float)
dim_max = np.array([60, 30, 100, 5], dtype = float)

assert employee_a.shape == employee_b.shape == dim_min.shape == dim_max.shape, \
    "All vectors must have the same number of dimensions"

# ── Min-max normalisation ─────────────────────────────────────────────────────
def minmax_normalise(v, v_min, v_max):
    return (v - v_min) / (v_max - v_min)

norm_a = minmax_normalise(employee_a, dim_min, dim_max)
norm_b = minmax_normalise(employee_b, dim_min, dim_max)

print(f"Employee A (raw):        {employee_a}")
print(f"Employee A (normalised): {norm_a.round(4)}")
print(f"\nEmployee B (raw):        {employee_b}")
print(f"Employee B (normalised): {norm_b.round(4)}")

# ── Euclidean distance before and after normalisation ─────────────────────────
dist_raw  = np.linalg.norm(employee_a - employee_b)
dist_norm = np.linalg.norm(norm_a - norm_b)

print(f"\nEuclidean distance (raw):        {dist_raw:.4f}")
print(f"  (dominated by engagement gap of {abs(employee_a[2] - employee_b[2]):.0f} points)")
print(f"\nEuclidean distance (normalised): {dist_norm:.4f}")
print(f"  (all dimensions now contribute proportionally to their plausible range)")

# ── Weighted normalised profile and composite score ───────────────────────────
# Deliberate analytical weights: performance matters most, age matters least
weights    = np.array([0.1, 0.2, 0.3, 0.4], dtype = float)
weighted_a = norm_a * weights
weighted_b = norm_b * weights

score_a       = weighted_a.sum()
score_b       = weighted_b.sum()
dist_weighted = np.linalg.norm(weighted_a - weighted_b)

print(f"\nWeights applied:                 {weights}")
print(f"  (age 0.1, tenure 0.2, engagement 0.3, performance 0.4)")
print(f"\nComposite score — Employee A:    {score_a:.4f}")
print(f"Composite score — Employee B:    {score_b:.4f}")
print(f"\nEuclidean distance (weighted):   {dist_weighted:.4f}")
print(f"  (distance now reflects analytical priorities, not just raw scale)")

# ── Visualisation — raw vs normalised for Employee A ─────────────────────────
x = np.arange(len(labels))

fig, axes = plt.subplots(
    1, 2,
    figsize   = (12, 5),
    facecolor = BG
)

for ax in axes:
    ax.set_facecolor(BG)
    ax.spines['bottom'].set_color(AXIS)
    ax.spines['left'].set_color(AXIS)
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.yaxis.grid(
        True,
        color     = GRID,
        linewidth = 0.6
    )
    ax.set_axisbelow(True)
    ax.tick_params(
        colors     = MUTED,
        labelcolor = SUB
    )

axes[0].bar(
    x, employee_a,
    width = 0.5,
    color = GRN
)
axes[0].set_xticks(x)
axes[0].set_xticklabels(
    labels,
    color    = SUB,
    fontsize = 10
)
axes[0].set_title(
    "Employee A — raw values (mixed scales)",
    color    = TEXT,
    fontsize = 10,
    pad      = 12
)

axes[1].bar(
    x, norm_a,
    width = 0.5,
    color = AMB
)
axes[1].set_xticks(x)
axes[1].set_xticklabels(
    labels,
    color    = SUB,
    fontsize = 10
)
axes[1].set_ylim(0, 1)
axes[1].set_title(
    "Employee A — normalised values (0 to 1)",
    color    = TEXT,
    fontsize = 10,
    pad      = 12
)

plt.suptitle(
    "Effect of min-max normalisation on a mixed-scale employee profile",
    color    = TEXT,
    fontsize = 11,
    y        = 1.02
)

plt.tight_layout()
plt.show()
```
