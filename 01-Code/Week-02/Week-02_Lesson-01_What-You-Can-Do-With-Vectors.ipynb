<h3>Three things I can do with a vector</h3>
<p>In Week 1, I learned what a vector is and how to measure its size. Now I want to know what I can actually do with one. It turns out there are three fundamental operations, and they are all simpler than they sound.</p>
<p><span class="term">Scalar multiplication</span> is scaling a vector. If I multiply a vector by a number, every element in that vector is multiplied by that same number. The shape of the vector stays the same, but its magnitude changes. Multiplying by 2 doubles every value and doubles the magnitude. Multiplying by 0.5 halves every value and halves the magnitude. Mathematically, I can scale any vector. In practice, I still need to ask whether the result remains meaningful when the vector mixes units or scales. In applied terms, if I have a salary vector representing pay for four roles and I apply a 10% pay uplift, I am performing scalar multiplication on a vector where all elements share the same unit, and the result is directly interpretable.</p>
<p><span class="term">Vector addition</span> is combining two vectors element by element. The first element of one vector is added to the first element of the other, and so on. The two vectors must have the same number of dimensions for this to be valid. In applied terms, if I have an employee profile vector representing their state at the start of the year and I have a change vector representing what shifted during the annual review cycle, then adding the change vector to the original gives me their updated profile.</p>
<p><span class="term">Vector subtraction</span> follows the same logic in reverse. Subtracting one vector from another gives me a difference vector. Every element in the result tells me by how much that dimension changed. This is particularly useful for measuring the gap between two states, or between two profiles.</p>
<p>These three operations are the building blocks of almost everything else in linear algebra. Scalar multiplication and vector addition combine to form something called a <span class="term">linear combination</span> &mdash; one of the most important ideas I will encounter throughout this programme.</p>
<div class="inner-formal">
<div class="inner-formal-label">Formal Notation</div>
<p>For a vector <strong>v</strong> with elements v&#8321;, v&#8322;, ..., v&#8345; and a scalar c:</p>
<p>Scalar multiplication: &nbsp;&nbsp; c<strong>v</strong> = [cv&#8321;, cv&#8322;, ..., cv&#8345;]</p>
<p>For two vectors <strong>a</strong> and <strong>b</strong> of the same length n:</p>
<p>Vector addition: &nbsp;&nbsp; <strong>a</strong> + <strong>b</strong> = [a&#8321; + b&#8321;, a&#8322; + b&#8322;, ..., a&#8345; + b&#8345;]</p>
<p>Vector subtraction: &nbsp;&nbsp; <strong>a</strong> &#8722; <strong>b</strong> = [a&#8321; &#8722; b&#8321;, a&#8322; &#8722; b&#8322;, ..., a&#8345; &#8722; b&#8345;]</p>
</div>
<div class="inner-note">
<div class="inner-note-label">Note on Magnitude</div>
<p>Scalar multiplication changes the magnitude of a vector. If I scale a vector by a factor c, the magnitude becomes |c| times the original magnitude. Vector addition and subtraction also change the magnitude, but not in a simple proportional way &mdash; the new magnitude depends on both the sizes of the vectors and the angle between them.</p>
</div>

<h3>Linear combinations at the heart of neural networks</h3>
<p><span class="term">Linear combinations</span>, built from scalar multiplication and vector addition, sit at the absolute centre of how neural networks work. Every neuron in a neural network takes a set of inputs, multiplies each one by a learned weight (scalar multiplication), sums the results (vector addition), adds a bias term, and passes the result through an activation function. The network learns by adjusting those weights across millions of training steps. When I hear that a large language model has billions of parameters, those parameters are overwhelmingly the weights in linear combinations being applied across vast numbers of vectors simultaneously. The operations introduced in this lesson are not preliminary concepts before the real AI work begins. They are the real AI work.</p>
</div>

<h3>Scaling, updating, and comparing employee profiles</h3>
<p>Three direct applications worth noting.</p>
<p>First, <span class="term">pay modelling</span>. A salary vector across a team of employees can be scaled by a single uplift factor to project the cost of a pay increase. The vector structure preserves the relative distribution while applying the change uniformly &mdash; and because all elements share the same unit, the scaled result is directly meaningful.</p>
<p>Second, <span class="term">profile updating</span>. Employee profiles in an HR system can be thought of as vectors in a feature space. An annual review process produces a change vector, and adding it to the current profile produces the updated one.</p>
<p>Third, <span class="term">gap analysis</span>. Subtracting one profile vector from another gives a difference vector. Each element of that difference tells me precisely how far apart two employees are on one dimension. That is the foundation for distance-based clustering and similarity analysis, which appear throughout workforce segmentation work.</p>
</div>

</div>


```python
import numpy             as np
import matplotlib.pyplot as plt

# ── Colour constants ──────────────────────────────────────────────────────────
BG    = '#0a0a0c'
TEXT  = '#e8e4de'
SUB   = '#b5b0a8'
MUTED = '#6a6560'
GRID  = '#2a2a28'
AXIS  = '#3a3a38'
GRN   = '#c4f042'
AMB   = '#f0a830'
RED   = '#e8453c'

# ── Employee profile ──────────────────────────────────────────────────────────
# Four dimensions: age, tenure (years), engagement score (0-100), performance (1-5)
profile       = np.array([34, 5, 72, 4], dtype = float)
change_vector = np.array([1, 1, 6, 0],  dtype = float)
labels        = ['Age', 'Tenure', 'Engagement', 'Performance']

assert profile.shape == change_vector.shape, \
    "Profile and change vector must match in shape"

print(f"Original profile: {profile}")
print(f"Dimensions:       {profile.shape[0]}")

# ── Scalar multiplication ─────────────────────────────────────────────────────
# Mathematical illustration only — scaling a mixed-unit vector by 0.5
# preserves the vector's shape but the result is not a meaningful employee profile
scale_factor   = 0.5
scaled_profile = scale_factor * profile

print(f"\nScale factor:     {scale_factor}")
print(f"Scaled profile:   {scaled_profile}")
print(f"  (mathematical illustration: halves every element proportionally)")
print(f"  (note: result is not interpretable as an employee profile — dimensions mix units)")

# ── Annual review change vector ───────────────────────────────────────────────
# One year older, one more year of tenure, +6 engagement, same performance rating
print(f"\nChange vector:    {change_vector}")

# ── Vector addition — updated profile ────────────────────────────────────────
updated_profile = profile + change_vector

print(f"Updated profile:  {updated_profile}")

# ── Vector subtraction — difference ──────────────────────────────────────────
difference = updated_profile - profile

print(f"\nDifference vector: {difference}")
print(f"  (should match the change vector exactly)")

# ── Magnitudes ────────────────────────────────────────────────────────────────
mag_original = np.linalg.norm(profile)
mag_updated  = np.linalg.norm(updated_profile)

print(f"\nMagnitude (original): {mag_original:.4f}")
print(f"Magnitude (updated):  {mag_updated:.4f}")

# ── Visualisation — grouped bar chart ────────────────────────────────────────
# Dimensions are on different scales — comparison is element by element only
print(f"\n(note: the chart below shows raw values across mixed-scale dimensions)")
print(f"  (engagement dominates the visual range — this is expected, not a plotting error)")

x         = np.arange(len(labels))
bar_width = 0.35

fig, ax = plt.subplots(
    figsize   = (9, 5),
    facecolor = BG
)
ax.set_facecolor(BG)

ax.bar(
    x - bar_width / 2,
    profile,
    width = bar_width,
    color = GRN,
    label = 'Original profile'
)
ax.bar(
    x + bar_width / 2,
    updated_profile,
    width = bar_width,
    color = AMB,
    label = 'Updated profile'
)

ax.set_xticks(x)
ax.set_xticklabels(
    labels,
    color    = SUB,
    fontsize = 10
)
ax.set_ylabel(
    "Value",
    color    = SUB,
    fontsize = 10
)
ax.set_title(
    "Raw employee profile values — before and after annual review (mixed scales)",
    color    = TEXT,
    fontsize = 11,
    pad      = 14
)
ax.tick_params(
    colors     = MUTED,
    labelcolor = SUB
)
ax.spines['bottom'].set_color(AXIS)
ax.spines['left'].set_color(AXIS)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.yaxis.grid(
    True,
    color     = GRID,
    linewidth = 0.6
)
ax.set_axisbelow(True)
ax.legend(
    facecolor  = BG,
    edgecolor  = AXIS,
    labelcolor = SUB,
    fontsize   = 9
)

plt.tight_layout()
plt.show()
```
