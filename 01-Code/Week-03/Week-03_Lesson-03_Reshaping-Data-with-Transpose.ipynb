<h3>Reshaping Data with Transpose</h3>
<p>I now know how to build a matrix, navigate it, and apply two operations to it. This lesson introduces a third operation that works differently from addition and scalar multiplication. It does not combine two matrices, and it does not scale values. It reorients a single matrix: flipping its rows into columns and its columns into rows.</p>
<p>This operation is called the <span class="term">transpose</span>.</p>
<h3>What transpose does</h3>
<p>Given a matrix <strong>A</strong>, its transpose is written <strong>A</strong><sup>T</sup>. The rule is straightforward: the element that was in row i, column j of <strong>A</strong> moves to row j, column i of <strong>A</strong><sup>T</sup>. Every row of the original becomes a column of the transpose, and every column of the original becomes a row.</p>
<p>The values in the matrix do not change. The data is identical before and after the operation. What changes is the orientation: which dimension runs across the rows and which runs down the columns.</p>
<p>The shape also changes. If <strong>A</strong> has shape m x n, then <strong>A</strong><sup>T</sup> has shape n x m. A (5 x 3) matrix becomes a (3 x 5) matrix after transposing. A (1 x 4) row vector becomes a (4 x 1) column vector. The only case where the shape does not change is a square matrix, where m and n are already equal.</p>
<div class="inner-formal">
<div class="inner-formal-label">Formal Notation</div>
<p>If <strong>A</strong> is an m x n matrix, then its transpose <strong>A</strong><sup>T</sup> is an n x m matrix defined by:</p>
<p>(<strong>A</strong><sup>T</sup>)<sub>ij</sub> = A<sub>ji</sub></p>
<p>The element in row i, column j of the transpose equals the element in row j, column i of the original. In NumPy, transpose is available as <code class="inline">A.T</code> or <code class="inline">np.transpose(A)</code>.</p>
</div>
<h3>What transpose does not do</h3>
<p>Transpose changes orientation. It does not summarise, aggregate, or compute anything. It does not add rows, average columns, or transform values in any way. The point of transposing is to reframe which dimension is most natural to work across, not to produce a new quantity from the original data. What changes after transposing is which operations become more natural to carry out. The transpose itself is not performing them.</p>
</div>
<div class="panel">
<div class="accent-strip accent-amber"></div>
<div class="sec-row">
<div class="sec-label amber">AI Connection</div>
<div class="sec-rule"></div>
</div>
<p>Transpose appears throughout linear algebra and machine learning in ways that are easy to overlook because it is often embedded inside larger operations. In the normal equation for linear regression, the closed-form solution involves <strong>X</strong><sup>T</sup><strong>X</strong> and <strong>X</strong><sup>T</sup><strong>y</strong>, where <strong>X</strong> is the feature matrix. The transpose is what allows the row-oriented data structure to be combined with itself in the right dimensional arrangement.</p>
<p>In neural networks, transpose is used when gradients need to flow backwards through a weight matrix during backpropagation. The forward pass multiplies by the weight matrix; the backward pass requires its transpose to propagate the error signal correctly. Understanding why that is the case becomes clearer once matrix multiplication is introduced in Week 4. For now, the important point is that transpose is not a niche reshaping convenience. It is embedded in the mechanics of how models learn.</p>
</div>
<div class="panel">
<div class="accent-strip accent-amber"></div>
<div class="sec-row">
<div class="sec-label amber">Workforce Analytics</div>
<div class="sec-rule"></div>
</div>
<p>Workforce datasets almost always arrive in employee-by-attribute orientation: one row per person, one column per metric. That orientation is natural for looking up an individual's full profile, or for filtering and selecting subsets of employees. It is the shape most analytical tools expect as input.</p>
<p>After transposing, the matrix becomes attribute-by-employee: one row per metric, one column per person. That orientation is more natural when the question is about a single variable across the whole workforce. If I want to inspect the distribution of engagement scores across all employees, or inspect two attributes side by side across the same employees, the transposed view puts the variable of interest into a row rather than a column, which can make certain operations more direct.</p>
<p>Neither orientation is inherently correct. The same data is present in both. The question is which frame makes the next analytical step easier to carry out and easier to read. In practice, data preparation for machine learning often involves transposing or reshaping matrices specifically to match the orientation a given function or model expects. Knowing how to reframe data without changing it is a core skill in applied data work.</p>
</div>
<div class="panel">
<div class="accent-strip accent-green"></div>
<div class="sec-row">
<div class="sec-label green">Worked Example</div>
<div class="sec-rule"></div>
</div>
<p>An employee-by-attribute matrix holds five employees across four attributes: tenure in years, performance score, absence days, and engagement score. The code below transposes it, confirms what changes and what does not, and demonstrates what becomes more natural to inspect in each orientation.</p>
<p>The transposed view makes it natural to loop across attributes and inspect each one as a row. The values are identical to those in the original matrix. The orientation is what has changed.</p>
</div>
</div>

```python
import numpy as np

# ── Employee by attribute matrix ───────────────────────────────────────────────
# Rows    : employees A, B, C, D, E
# Columns : tenure (yrs) | performance (1-10) | absence (days) | engagement (1-10)

workforce = np.array([
    [3,  7,  4,  8],
    [7,  9,  2,  9],
    [1,  5, 11,  5],
    [5,  8,  6,  7],
    [4,  6,  8,  6],
], dtype = float)

# ── Shape before and after transpose ──────────────────────────────────────────
workforce_t = workforce.T

print(f"Original shape:   {workforce.shape}  (employees x attributes)")
print(f"Transposed shape: {workforce_t.shape}  (attributes x employees)")

# ── Values are unchanged: element check ───────────────────────────────────────
print(f"\nworkforce[2, 3]   (employee C, engagement): {workforce[2, 3]:.0f}")
print(f"workforce_t[3, 2] (engagement, employee C): {workforce_t[3, 2]:.0f}")

# ── Original orientation: one employee's full profile ─────────────────────────
print(f"\nEmployee D full profile (original row 3): {workforce[3]}")

# ── Transposed orientation: one attribute across all employees ─────────────────
attr_labels = ["Tenure", "Performance", "Absence", "Engagement"]

print(f"\nAll employees by attribute (transposed rows):")
for i, label in enumerate(attr_labels):
    print(f"  {label:12}: {workforce_t[i]}")

# ── Per-attribute summary in transposed view ───────────────────────────────────
print(f"\nAttribute means across all employees:")
for i, label in enumerate(attr_labels):
    print(f"  {label:12}: {workforce_t[i].mean():.2f}")
```
