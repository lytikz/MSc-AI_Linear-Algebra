<h3>What a Transformation Actually Does</h3>
<p>In Lesson 1, the focus was on the mechanics of matrix multiplication: how to compute each element, what the shape rule is, and how to verify the result. The multiplication was treated as an arithmetic process.</p>
<p>In this lesson the question shifts. Instead of asking how to compute the result, I am asking what the result means. What has actually happened to the data after multiplication? What did the matrix do?</p>
<p>When I multiply a matrix of employee vectors by a transformation matrix, each employee's profile is mapped to a new set of values. The transformation matrix defines the rule. Every employee goes through the same rule. The result is a new matrix where every employee now lives in a different space, described by different dimensions.</p>
<p>A simple and important example is scaling. If the transformation matrix scales one attribute by a factor of 2 and leaves everything else unchanged, then after multiplication every employee's value on that attribute is doubled. The matrix encoded the instruction, and multiplication applied it uniformly across all employees.</p>
<p>A more interesting example is rotation. A rotation matrix in 2D takes a set of vectors pointing in various directions and rotates all of them by the same angle. The vectors change direction but their lengths are preserved. The transformation is the rotation, and the matrix encodes it.</p>
<p>In both cases the key insight is the same. A matrix is not just a container for numbers. When used as the right-hand operand in a multiplication, it is a function. It takes every row of the left-hand matrix and maps it to a new location according to the same rule.</p>
<p>For a dataset <strong>X</strong> of shape (n, k), the transformation matrix must have k rows so each input dimension can contribute to the output. Each column of the transformation matrix defines one output dimension as a weighted combination of the input dimensions. If the transformation matrix has m columns, the result has shape (n, m): the same employees, now described in a new space with m dimensions.</p>
<p>One critical property carries over from Lesson 1: matrix multiplication is not <span class="term">commutative</span>. AB applies the recipe in B to the rows of A. BA applies the recipe in A to the rows of B. Those are different operations on different objects. In general they produce different results, and often one of them will not even have valid dimensions.</p>
<div class="inner-formal">
<div class="inner-formal-label">Formal Notation</div>
<p>A transformation can be written as <strong>y</strong> = <strong>Ax</strong>, where <strong>x</strong> is an input column vector, <strong>A</strong> is the transformation matrix, and <strong>y</strong> is the output vector. When applied across an entire dataset stored as row vectors, the employee matrix <strong>X</strong> of shape (n, k) is transformed by <strong>A</strong> of shape (k, m) to produce <strong>Y</strong> = <strong>XA</strong> of shape (n, m). Each row of <strong>Y</strong> is the transformed version of the corresponding row of <strong>X</strong>. The transformation is <span class="term">linear</span>: scaling an input scales the output by the same factor, and transforming a sum of inputs gives the same result as summing the transformed inputs separately.</p>
</div>
</div>
<div class="panel">
<div class="accent-strip accent-amber"></div>
<div class="sec-row">
<div class="sec-label amber">AI Connection</div>
<div class="sec-rule"></div>
</div>
<p>The transformation framing is how neural networks are understood mathematically. Each layer takes an input matrix, applies a weight matrix as a transformation, adds a bias, and passes the result through a non-linear activation function. The sequence of layers is a sequence of transformations, each one reshaping the data into a new space where the next transformation can operate.</p>
<p>The reason deep networks are powerful is that a composition of simple linear transformations, interleaved with non-linearities, can approximate extraordinarily complex functions. But every individual step is a matrix transformation of exactly the type covered in this lesson. PCA, which appears in Week 11, is also a transformation: it finds the directions in which data varies most and transforms the dataset into a new coordinate system aligned with those directions.</p>
</div>
<div class="panel">
<div class="accent-strip accent-amber"></div>
<div class="sec-row">
<div class="sec-label amber">Workforce Analytics</div>
<div class="sec-rule"></div>
</div>
<p>A common task in people analytics is scaling a dataset so that attributes on different units can be compared directly. After centring the data by subtracting column means, a diagonal scaling matrix can then scale each attribute by the reciprocal of its standard deviation. That scaling step is a matrix transformation: every employee goes through the same recipe, and the result is a new matrix where the columns have unit variance.</p>
<p>The same logic applies to more complex operations: projecting a high-dimensional employee dataset down to two dimensions for visualisation, rotating a feature space to align with principal components, or mapping raw attributes to composite scores. In each case, a matrix encodes the rule and multiplication applies it uniformly across all employees.</p>

<div class="inner-note-label">Professional Note</div>
<p>The transformation example in the worked example below uses a 2D geometric rotation to make the mechanics visible. This is deliberately simpler than a workforce example so that the effect of the transformation can be seen clearly on a plot. The workforce connection above describes how the same principle applies in real people analytics work.</p>

<div class="panel">
<div class="accent-strip accent-green"></div>
<div class="sec-row">
<div class="sec-label green">Worked Example</div>
<div class="sec-rule"></div>
</div>
<p>Six employee vectors, each described by two attributes: standardised performance score and standardised engagement score. Both are on a common scale so the 2D geometry is meaningful.</p>
<p>The transformation matrix rotates all vectors by 45 degrees anticlockwise. The code works through the rotation matrix construction, applies the transformation, and then verifies two properties: that vector lengths are preserved, and that the matrix is a rotation rather than a reflection. The original and transformed vectors are shown in side by side plots so the effect of the transformation is visible directly.</p>
</div>
</div>

```python
import numpy             as np
import matplotlib.pyplot as plt

# ── Colour constants ───────────────────────────────────────────────────────────
BG    = '#0a0a0c'
TEXT  = '#e8e4de'
SUB   = '#b5b0a8'
MUTED = '#6a6560'
GRID  = '#2a2a28'
AXIS  = '#3a3a38'
GRN   = '#c4f042'

# ── Employee vectors (6 employees x 2 attributes) ─────────────────────────────
# Columns : standardised performance | standardised engagement
# Row vector convention — employees @ R, not R @ employees

employees = np.array([
    [ 1.2,  0.5],
    [ 0.3,  1.4],
    [-0.8,  0.9],
    [-1.1, -0.4],
    [ 0.6, -1.2],
    [ 1.5,  1.1],
], dtype = float)

names = ['A', 'B', 'C', 'D', 'E', 'F']

# ── Rotation matrix: 45 degrees anticlockwise for row vectors ─────────────────
# For row vectors right-multiplied by R, anticlockwise rotation uses:
#   [ cos   sin]
#   [-sin   cos]
# This is the transpose of the standard column-vector form.

theta = np.radians(45)
R     = np.array([
    [ np.cos(theta), np.sin(theta)],
    [-np.sin(theta), np.cos(theta)],
], dtype = float)

print(f"Employee matrix shape:  {employees.shape}")
print(f"Rotation matrix shape:  {R.shape}")
print(f"Result shape will be:   {employees.shape[0]} x {R.shape[1]}")

# ── Apply transformation ───────────────────────────────────────────────────────
transformed = employees @ R

print(f"\nOriginal vs transformed positions:")
print(f"{'Employee':<10} {'Original':>20} {'Transformed':>22}")
print(f"{'-' * 54}")
for i, name in enumerate(names):
    orig  = f"[{employees[i,0]:+.2f}, {employees[i,1]:+.2f}]"
    trans = f"[{transformed[i,0]:+.2f}, {transformed[i,1]:+.2f}]"
    print(f"{name:<10} {orig:>20} {trans:>22}")

# ── Confirm lengths are preserved ─────────────────────────────────────────────
orig_lengths  = np.linalg.norm(employees,   axis = 1)
trans_lengths = np.linalg.norm(transformed, axis = 1)

print(f"\nLength preservation check:")
print(f"{'Employee':<10} {'Before':>10} {'After':>10} {'Match':>8}")
print(f"{'-' * 40}")
for i, name in enumerate(names):
    match = np.isclose(orig_lengths[i], trans_lengths[i])
    print(f"{name:<10} {orig_lengths[i]:>10.4f} {trans_lengths[i]:>10.4f} {str(match):>8}")

# ── Confirm rotation (not just orthogonality) ──────────────────────────────────
# R @ R.T = I confirms the transformation is orthogonal (lengths and angles preserved).
# det(R) = +1 confirms it is a rotation rather than a reflection.

det_R = np.linalg.det(R)

print(f"\nR @ R.T gives identity:  {np.allclose(R @ R.T, np.eye(2))}")
print(f"det(R) = +1 (rotation):  {np.isclose(det_R, 1.0)}")
print(f"  det(R) value:          {det_R:.6f}")

# ── Side by side plots ─────────────────────────────────────────────────────────
fig, axes = plt.subplots(
    1, 2,
    figsize   = (12, 5.5),
    facecolor = BG
)

for ax, data, title in zip(
    axes,
    [employees, transformed],
    ['Original', 'After 45-degree anticlockwise rotation']
):
    ax.set_facecolor(BG)
    ax.set_aspect('equal', adjustable = 'box')
    ax.axhline(0, color = AXIS, linewidth = 0.8)
    ax.axvline(0, color = AXIS, linewidth = 0.8)
    ax.grid(
        True,
        color     = GRID,
        linewidth = 0.5,
        linestyle = '--'
    )

    for i, name in enumerate(names):
        ax.quiver(
            0, 0, data[i, 0], data[i, 1],
            angles      = 'xy',
            scale_units = 'xy',
            scale       = 1,
            color       = GRN,
            width       = 0.012,
            alpha       = 0.85
        )
        ax.annotate(
            name,
            xy     = (data[i, 0], data[i, 1]),
            xytext = (data[i, 0] + 0.06, data[i, 1] + 0.06),
            color  = TEXT,
            fontsize = 9
        )

    ax.set_title(
        title,
        color    = TEXT,
        fontsize = 11,
        pad      = 14
    )
    ax.set_xlabel("Performance (standardised)", color = SUB, fontsize = 9)
    ax.set_ylabel("Engagement (standardised)",  color = SUB, fontsize = 9)
    ax.tick_params(colors = MUTED)
    for spine in ax.spines.values():
        spine.set_edgecolor(AXIS)

plt.suptitle(
    "Effect of a 45-degree anticlockwise rotation on employee vectors",
    color    = TEXT,
    fontsize = 12,
    y        = 1.02
)
plt.tight_layout()
plt.show()
```
