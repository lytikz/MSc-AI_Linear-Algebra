<h3>From element-by-element gaps to a single distance</h3>
<p>In Lesson 1, I learned how to update a vector and measure the gap between two vectors element by element. But a difference vector gives me four numbers, one per dimension. Sometimes I want a single number that summarises how far apart two vectors are overall. That is what distance gives me.</p>
<p>The most natural way to measure distance between two vectors is <span class="term">Euclidean distance</span>. The idea comes directly from geometry. If I have two points in 2D space, the distance between them is the length of the straight line connecting them &mdash; what I would measure with a ruler. In higher dimensions, the same principle extends exactly: I compute the difference vector, square each element, sum the squares, and take the square root. That final number is the Euclidean distance.</p>
<p>This connects back to something I already know. The <span class="term">magnitude</span> of a vector is its Euclidean distance from the origin &mdash; from the zero vector. Euclidean distance between two vectors <strong>a</strong> and <strong>b</strong> is simply the magnitude of their difference: &#8214;<strong>a</strong> &#8722; <strong>b</strong>&#8214;. I am not learning something new so much as applying what I already have to a pair of points rather than a single one.</p>
<p>There is an important distinction to carry forward from Week 1. Cosine similarity asks: do these two vectors point in the same direction? Euclidean distance asks: how far apart are these two vectors in space? These are different questions and they can give different answers. Two employees could have very similar engagement and performance patterns relative to their own scales &mdash; high cosine similarity &mdash; but sit far apart in raw Euclidean distance simply because one has much larger absolute values. Both measures are valid. They answer different questions. Knowing which question I am actually asking is most of the work.</p>
<p>As with scalar multiplication in Lesson 1, Euclidean distance on a mixed-unit vector needs to be treated with care. If one dimension is measured on a 0 to 100 scale and another on a 1 to 5 scale, the large-scale dimension will dominate the distance calculation. This is not a flaw in the mathematics. It is a signal that normalisation &mdash; the subject of Lesson 3 &mdash; matters.</p>
<div class="inner-formal">
<div class="inner-formal-label">Formal Notation</div>
<p>For two vectors <strong>a</strong> and <strong>b</strong> of the same length n, the Euclidean distance is:</p>
<p>d(<strong>a</strong>, <strong>b</strong>) = &#8214;<strong>a</strong> &#8722; <strong>b</strong>&#8214; = &#8730;[(a&#8321; &#8722; b&#8321;)&#178; + (a&#8322; &#8722; b&#8322;)&#178; + ... + (a&#8345; &#8722; b&#8345;)&#178;]</p>
<p>In NumPy this is computed directly as <code class="inline">np.linalg.norm(a - b)</code>.</p>

<h3>Distance at the heart of learning algorithms</h3>
<p>Distance measures are everywhere in machine learning. The <span class="term">k-nearest neighbours</span> algorithm classifies a new data point by finding the k training points with the smallest Euclidean distance to it and taking a majority vote. Clustering methods such as <span class="term">k-means</span> assign points to the nearest cluster centre and update those centres by averaging the vectors in each cluster &mdash; under the hood, standard k-means is optimising squared Euclidean distance. In embedding spaces, where words, sentences, or user behaviours are represented as high-dimensional vectors, Euclidean distance and cosine similarity are both used depending on whether the absolute position in space or the direction of the vector matters more for the task at hand. The choice between them is not arbitrary &mdash; it reflects a deliberate decision about what similarity means in that context.</p>
</div>

<h3>Proximity-based scoring and segmentation</h3>
<p>Distance-based thinking appears throughout workforce analytics in ways that are not always labelled as such. Prototype-based scoring can compare a current employee's profile with a reference profile of past leavers or other target groups to produce a proximity-based risk indicator. Succession planning tools can compute the distance between a candidate's profile vector and a target role profile to identify development gaps. Employee clustering for benefits segmentation or communication targeting groups people by proximity in feature space. In all of these cases, the raw distance number is rarely the final output. What matters is the relative ranking &mdash; which employees are closest to which profiles &mdash; and whether the dimensions have been put on comparable scales before the distance is computed.</p>
</div>

</div>


```python
import numpy             as np
import matplotlib.pyplot as plt

# ── Colour constants ──────────────────────────────────────────────────────────
BG    = '#0a0a0c'
TEXT  = '#e8e4de'
SUB   = '#b5b0a8'
MUTED = '#6a6560'
GRID  = '#2a2a28'
AXIS  = '#3a3a38'
GRN   = '#c4f042'
AMB   = '#f0a830'
RED   = '#e8453c'

# ── Two employee profiles ─────────────────────────────────────────────────────
# Four dimensions: age, tenure (years), engagement score (0-100), performance (1-5)
employee_a = np.array([34,  5, 72, 4], dtype = float)
employee_b = np.array([29, 11, 58, 3], dtype = float)
labels     = ['Age', 'Tenure', 'Engagement', 'Performance']

assert employee_a.shape == employee_b.shape, \
    "Both employee vectors must have the same number of dimensions"
assert np.any(employee_a != 0) and np.any(employee_b != 0), \
    "Employee profile vectors must not be zero vectors"

print(f"Employee A: {employee_a}")
print(f"Employee B: {employee_b}")

# ── Difference vector ─────────────────────────────────────────────────────────
difference = employee_a - employee_b

print(f"\nDifference vector (A - B): {difference}")
print(f"  (each element: how much A exceeds B on that dimension)")

# ── Euclidean distance ────────────────────────────────────────────────────────
euclidean_distance = np.linalg.norm(employee_a - employee_b)

print(f"\nEuclidean distance:        {euclidean_distance:.4f}")
print(f"  (note: dominated by engagement gap — dimensions are on different scales)")

# ── Cosine similarity ─────────────────────────────────────────────────────────
def cosine_similarity(a, b):
    return np.dot(a, b) / (np.linalg.norm(a) * np.linalg.norm(b))

cos_sim = cosine_similarity(employee_a, employee_b)

print(f"\nCosine similarity:         {cos_sim:.4f}")
print(f"  (close to 1.0 — both profiles point in broadly the same direction)")
print(f"  (high similarity does not mean small distance — these measure different things)")

# ── Visualisation — profiles side by side ────────────────────────────────────
# Distance appears in the chart title — dimensions are on different scales
print(f"\n(note: chart shows raw values across mixed-scale dimensions)")
print(f"  (engagement dominates the visual range and the distance calculation)")

x         = np.arange(len(labels))
bar_width = 0.35

fig, ax = plt.subplots(
    figsize   = (9, 5),
    facecolor = BG
)
ax.set_facecolor(BG)

ax.bar(
    x - bar_width / 2,
    employee_a,
    width = bar_width,
    color = GRN,
    label = 'Employee A'
)
ax.bar(
    x + bar_width / 2,
    employee_b,
    width = bar_width,
    color = AMB,
    label = 'Employee B'
)

ax.set_xticks(x)
ax.set_xticklabels(
    labels,
    color    = SUB,
    fontsize = 10
)
ax.set_ylabel(
    "Value",
    color    = SUB,
    fontsize = 10
)
ax.set_title(
    f"Two employee profiles — Euclidean distance: {euclidean_distance:.2f} (mixed scales)",
    color    = TEXT,
    fontsize = 11,
    pad      = 14
)
ax.tick_params(
    colors     = MUTED,
    labelcolor = SUB
)
ax.spines['bottom'].set_color(AXIS)
ax.spines['left'].set_color(AXIS)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.yaxis.grid(
    True,
    color     = GRID,
    linewidth = 0.6
)
ax.set_axisbelow(True)
ax.legend(
    facecolor  = BG,
    edgecolor  = AXIS,
    labelcolor = SUB,
    fontsize   = 9
)

plt.tight_layout()
plt.show()
```
