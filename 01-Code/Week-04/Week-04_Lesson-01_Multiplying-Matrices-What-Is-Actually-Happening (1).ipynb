<h3>What Matrix Multiplication Is</h3>
<p>The first thing to be clear about is what matrix multiplication is not. It is not elementwise. Adding two matrices together works by pairing up each element in the same position and adding them. Matrix multiplication does something fundamentally different.</p>
<p>To multiply two matrices, I take one row from the left-hand matrix and one column from the right-hand matrix, multiply the corresponding values together position by position, and sum all the results. That sum is a <span class="term">dot product</span>. I repeat that process for every row-column combination to fill out the result matrix.</p>
<p>Each element in the result is computed from exactly one row of the left matrix and exactly one column of the right matrix. The element at position (i, j) in the result is the dot product of row i from the left matrix and column j from the right matrix.</p>
<p>This has an immediate consequence for shapes. If the left matrix is m by k, the right matrix must have k rows. Every row of the left matrix has k values, and every column of the right matrix must also have k values to make the dot products possible. If the right matrix is k by n, the result will be m by n. The inner dimension k must agree, and then disappears. The outer dimensions m and n survive into the result.</p>
<p>The shape rule is: (m by k) times (k by n) produces (m by n).</p>
<p>A useful way to read this in a workforce context: if the left matrix has employees as rows and k attributes as columns, the right matrix must have k rows. The result has the same employees in its rows, and its columns are whatever the right matrix was producing as outputs. Matrix multiplication transforms a set of inputs into a set of outputs, one employee at a time, using a consistent recipe defined by the right matrix.</p>
<p>One critical property: matrix multiplication is not <span class="term">commutative</span>. AB and BA are generally not equal, and in many cases one of them will not even be a valid operation because the shapes will not allow it.</p>
<div class="inner-formal">
<div class="inner-formal-label">Formal Notation</div>
<p>A matrix is written as a bold capital letter such as <strong>A</strong>. If <strong>A</strong> has shape m by k and <strong>B</strong> has shape k by n, their product <strong>C</strong> = <strong>AB</strong> has shape m by n. The element at row i, column j in <strong>C</strong> is:</p>
<p>C<sub>ij</sub> = &sum;<sub>s=1</sub><sup>k</sup> A<sub>is</sub> B<sub>sj</sub></p>
<p>This sums k pairwise products across the shared inner dimension. Mathematical notation counts rows and columns from 1. NumPy uses zero-based indexing, so the element at mathematical position (i, j) is accessed as <code class="inline">C[i-1, j-1]</code> in code — a direct callback to the indexing rule from Week 3. The matrix multiplication operator in NumPy is <code class="inline">@</code>. Using <code class="inline">*</code> instead performs elementwise multiplication, which is a different operation entirely.</p>
</div>
<h3>The Identity Matrix</h3>
<p>There is one special matrix worth knowing now. The <span class="term">identity matrix</span> is a square matrix with 1s along the main diagonal and 0s everywhere else. Multiplying any matrix by a compatible identity matrix leaves it completely unchanged. It plays exactly the same role for matrices as the number 1 plays in ordinary multiplication. It reappears throughout the rest of the programme and serves as a useful sanity check in code.</p>
<div class="inner-card">
<div class="inner-card-label">Perspective</div>
<p>The identity matrix is the baseline transformation: apply it and nothing moves. Every more interesting matrix does something to the data it multiplies. Eigenvalues and eigenvectors, which arrive in Week 8, are defined precisely in terms of what a matrix does compared to what the identity would do. Getting comfortable with the identity now pays off later.</p>
</div>
</div>
<div class="panel">
<div class="accent-strip accent-amber"></div>
<div class="sec-row">
<div class="sec-label amber">AI Connection</div>
<div class="sec-rule"></div>
</div>
<p>At every layer of a neural network, the input data is multiplied by a <span class="term">weight matrix</span>. Each row of the input represents one example, and the multiplication transforms it into a new set of values that the next layer receives. In practice, each layer also adds a bias term and passes the result through a non-linear activation function, but the weight matrix multiplication is the central operation. Training a network is fundamentally the process of finding weight matrices whose successive multiplications produce useful outputs. The architecture of a network, how many layers it has and how wide each one is, defines the shapes of those weight matrices. Understanding matrix multiplication is understanding what a neural network does at every step.</p>
</div>
<div class="panel">
<div class="accent-strip accent-amber"></div>
<div class="sec-row">
<div class="sec-label amber">Workforce Analytics</div>
<div class="sec-rule"></div>
</div>
<p>A common people analytics task is computing composite scores from a set of raw attributes. If I have an employee-by-attribute matrix and want to collapse those attributes into a smaller number of summary indices, I can define a weighting matrix whose columns each represent one composite score recipe. Multiplying the employee matrix by the weighting matrix gives a result where each row is still one employee and each column is one composite score. The dot product mechanics ensure that every score is computed consistently for every employee using the same weighting recipe.</p>
<p>This is not just a convenient shorthand for doing the calculation column by column. It is the natural matrix framing of a transformation: take a set of inputs described in one space, map them to a new space using a defined recipe. Eigenvalues, SVD, and PCA in later weeks are all variations of this same idea.</p>
<div class="inner-note">
<div class="inner-note-label">Professional Note</div>
<p>The composite scores in the worked example are illustrative teaching scores, not production-grade indices. In real people analytics, combining attributes on different scales without first normalising them risks having one attribute dominate purely because of its numerical range. The normalisation techniques from Week 2 would apply before any real weighting exercise.</p>
</div>
</div>
<div class="panel">
<div class="accent-strip accent-green"></div>
<div class="sec-row">
<div class="sec-label green">Worked Example</div>
<div class="sec-rule"></div>
</div>
<p>Three employees, Alex, Blake, and Casey, each described by four attributes: tenure in years, performance score (1 to 10), absence days, and engagement score (1 to 10). The workforce matrix has shape (3, 4). A weighting matrix of shape (4, 2) maps those four attributes to two illustrative composite scores.</p>
<p>Before computing the full result, the code works through one element step by step: it multiplies the values pairwise, sums them by hand, and then confirms the same value with NumPy. The full result is then computed with <code class="inline">@</code>. A final identity matrix check confirms that multiplying by the identity leaves the original data unchanged.</p>
</div>
</div>

```python
import numpy as np

# ── Workforce matrix (3 employees x 4 attributes) ─────────────────────────────
# Rows    : Alex, Blake, Casey
# Columns : tenure (yrs) | performance (1-10) | absence (days) | engagement (1-10)

workforce = np.array([
    [4,  8,  3,  9],
    [2,  5, 11,  6],
    [7,  9,  1,  8],
], dtype = float)

# ── Weighting matrix (4 attributes x 2 composite scores) ──────────────────────
# Column 0 : Score A — illustrative retention-style weighting
# Column 1 : Score B — illustrative development-style weighting

weights = np.array([
    [-0.2,  0.1],
    [ 0.1,  0.5],
    [ 0.6, -0.1],
    [ 0.1,  0.4],
], dtype = float)

print(f"Workforce matrix shape:  {workforce.shape}")
print(f"Weighting matrix shape:  {weights.shape}")
print(f"Result shape will be:    ({workforce.shape[0]}, {weights.shape[1]})")

# ── One element by hand ────────────────────────────────────────────────────────
# Result[0, 0] is the dot product of workforce row 0 and weights column 0.
# Step 1: multiply corresponding values pairwise.
# Step 2: sum the products.

alex_row      = workforce[0]
score_a_col   = weights[:, 0]
pair_products = alex_row * score_a_col
manual_sum    = pair_products.sum()
numpy_check   = np.dot(alex_row, score_a_col)

print(f"\nAlex profile row:        {alex_row}")
print(f"Score A weight column:   {score_a_col}")
print(f"Pairwise products:       {pair_products}")
print(f"  = (4*-0.2) + (8*0.1) + (3*0.6) + (9*0.1)")
print(f"Sum by hand:             {manual_sum:.3f}")
print(f"Confirmed with np.dot:   {numpy_check:.3f}")

# ── Full matrix multiplication ─────────────────────────────────────────────────
scores = workforce @ weights

print(f"\nComposite scores (all employees):")
print(f"{'Employee':<10} {'Score A':>12} {'Score B':>12}")
print(f"{'-' * 36}")

names = ['Alex', 'Blake', 'Casey']
for i, name in enumerate(names):
    print(f"{name:<10} {scores[i, 0]:>12.3f} {scores[i, 1]:>12.3f}")

# ── Identity matrix check ──────────────────────────────────────────────────────
# A (3, 4) workforce matrix multiplied by a (4, 4) identity produces (3, 4).
# The result must be identical to the original.

identity_4      = np.eye(workforce.shape[1], dtype = float)
workforce_check = workforce @ identity_4

print(f"\nIdentity matrix shape:          {identity_4.shape}")
print(f"workforce @ I matches original: {np.allclose(workforce, workforce_check)}")
```
