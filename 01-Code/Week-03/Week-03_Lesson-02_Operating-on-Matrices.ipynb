<h3>Operating on Matrices</h3>
<p>Now that I know what a matrix is and how to navigate one, the next question is what I can actually do with one. In this lesson I will look at two operations: adding two matrices together, and multiplying a matrix by a scalar. Both are straightforward in mechanics, but each carries a rule that matters.</p>
<h3>Matrix addition</h3>
<p>If I have two matrices and I want to add them together, I add the element in position (i, j) of the first matrix to the element in position (i, j) of the second. The result is a new matrix of the same shape, where every position holds the sum of the two corresponding elements. This is called <span class="term">elementwise addition</span>.</p>
<p>The critical rule is <span class="term">shape compatibility</span>. Matrix addition is only defined when both matrices have exactly the same shape. I cannot add a (4 x 3) matrix to a (3 x 4) matrix. The dimensions must match in both directions. Beyond shape, the rows and columns of the two matrices must also carry the same meaning. If I am adding two headcount tables, both tables must have the same departments in the same row order and the same months in the same column order. A mismatch in either shape or meaning produces a result that is arithmetically valid but analytically meaningless.</p>
<p>Matrix addition is also <span class="term">commutative</span>: A + B gives the same result as B + A. The order in which I add two matrices does not affect the outcome, just as with ordinary addition of numbers.</p>
<div class="inner-formal">
<div class="inner-formal-label">Formal Notation</div>
<p>For two m x n matrices A and B, the sum C = A + B is defined as:</p>
<p>C<sub>ij</sub> = A<sub>ij</sub> + B<sub>ij</sub> for all i and j.</p>
<p>In NumPy, matrix addition is written with the standard <code class="inline">+</code> operator and applied elementwise automatically.</p>
</div>
<h3>Scalar multiplication</h3>
<p>Multiplying a matrix by a <span class="term">scalar</span> multiplies every element of the matrix by that single number. If I multiply a headcount table by 1.10, every element increases by 10 percent. If I multiply by 0.85, every element falls by 15 percent. The shape of the matrix does not change. Every position is scaled by the same factor.</p>
<p>The important interpretive point is that scalar multiplication is a modelling tool, not a description of what has already happened. Multiplying a headcount table by 1.10 produces a planning scenario showing what staffing would look like under a 10 percent growth assumption. It does not describe actual future headcount, and the resulting values may be fractional because the output is a planning model rather than a literal count of whole employees. This distinction matters in any analytical context where the output will inform a decision.</p>
<div class="inner-formal">
<div class="inner-formal-label">Formal Notation</div>
<p>For a scalar k and an m x n matrix A, the product kA is defined as:</p>
<p>(kA)<sub>ij</sub> = k x A<sub>ij</sub> for all i and j.</p>
<p>In NumPy, scalar multiplication uses the standard <code class="inline">*</code> operator.</p>
</div>
</div>
<div class="panel">
<div class="accent-strip accent-amber"></div>
<div class="sec-row">
<div class="sec-label amber">AI Connection</div>
<div class="sec-rule"></div>
</div>
<p>In machine learning, scalar multiplication appears whenever a learning rate is applied during training. The learning rate is a small scalar, typically something like 0.001, and it multiplies the gradient matrix at each update step. Without scalar multiplication, the model would update its weights by the raw gradient, which is almost always too large a step. Scaling by the learning rate controls how far the model moves at each iteration.</p>
<p>Matrix addition appears in how model outputs are combined. In ensemble methods, predictions from multiple models can be arranged in matching arrays and combined elementwise, often by averaging, to produce a final prediction. The shape compatibility rule applies directly: the prediction arrays must have identical structure for the combination to be valid.</p>
</div>
<div class="panel">
<div class="accent-strip accent-amber"></div>
<div class="sec-row">
<div class="sec-label amber">Workforce Analytics</div>
<div class="sec-rule"></div>
</div>
<p>Headcount reporting is where matrix addition becomes immediately practical. Imagine I am responsible for two business units, each of which produces a monthly headcount table with departments as rows and months as columns. Adding the two tables elementwise gives me a combined staffing picture across both units, with every cell representing total headcount in that department for that month. The result answers a business question directly: what is our combined staffing position?</p>
<p>Scalar multiplication is the natural tool for headcount scenario planning. If leadership asks what the workforce would look like under a 10 percent growth plan, I multiply the current headcount table by 1.10. If I need to model a 15 percent reduction, I multiply by 0.85. The resulting table is a planning scenario, not a forecast or a commitment, and the values it contains are planning figures rather than counts of actual people. That framing needs to be explicit whenever these outputs go to a business audience.</p>
<p>Both operations also appear together in workforce rebalancing exercises: scale one business unit's headcount to reflect a planned change, then add it to a second unit's table to see the combined effect on total staffing.</p>
</div>
<div class="panel">
<div class="accent-strip accent-green"></div>
<div class="sec-row">
<div class="sec-label green">Worked Example</div>
<div class="sec-rule"></div>
</div>
<p>Two business units each produce a three-month headcount table covering three departments. The code below adds them to produce a combined view, applies scalar multiplication to model a growth scenario, and verifies that addition is commutative.</p>
<p>The output confirms shape compatibility before the addition runs, shows the combined table, reports total headcount per month as a column summary, applies the growth scenario via scalar multiplication, and verifies commutativity. The scenario values are fractional because the output is a planning model rather than a count of whole employees.</p>
</div>
</div>

```python
import numpy as np

# ── Headcount matrices ─────────────────────────────────────────────────────────
# Rows    : departments (Engineering, Sales, Operations)
# Columns : months (January, February, March)

unit_a = np.array([
    [42, 44, 45],
    [18, 18, 20],
    [31, 30, 32],
], dtype = float)

unit_b = np.array([
    [28, 29, 30],
    [12, 14, 14],
    [22, 22, 24],
], dtype = float)

# ── Shape check before addition ────────────────────────────────────────────────
print(f"Unit A shape: {unit_a.shape}")
print(f"Unit B shape: {unit_b.shape}")
print(f"Compatible:   {unit_a.shape == unit_b.shape}")

# ── Matrix addition: combined headcount ────────────────────────────────────────
combined = unit_a + unit_b

print(f"\nCombined headcount (Jan to Mar):")
print(combined)

# ── Column totals: total headcount per month ───────────────────────────────────
print(f"\nTotal headcount by month:")
print(f"  January:  {combined[:, 0].sum():.0f}")
print(f"  February: {combined[:, 1].sum():.0f}")
print(f"  March:    {combined[:, 2].sum():.0f}")

# ── Scalar multiplication: 10 percent growth scenario ─────────────────────────
growth_factor = 1.10
scenario      = combined * growth_factor

print(f"\n10 percent growth scenario (planning figures, not literal headcount):")
print(np.round(scenario, 1))

# ── Commutativity check ────────────────────────────────────────────────────────
print(f"\nA + B equals B + A: {np.allclose(unit_a + unit_b, unit_b + unit_a)}")
```
