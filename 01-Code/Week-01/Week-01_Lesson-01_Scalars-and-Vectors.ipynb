<h3>What is a vector, exactly?</h3>
<p>A <span class="term">scalar</span> is just a single number. Temperature, salary, age, a test score &mdash; all scalars. One number, one piece of information. That is all.</p>
<p>A <span class="term">vector</span> is a collection of numbers arranged in order. The order matters. Think of an employee record: age 34, tenure 5 years, engagement score 72, performance rating 4. Those four numbers together form a <span class="term">four-dimensional vector</span>. Remove one value or shuffle the order and the meaning changes.</p>
<div class="inner-card">
<div class="inner-card-label">Perspective</div>
<p>A vector can be understood in two complementary ways, and I need both.</p>
<p>The <span class="term">geometric view</span> treats a vector as an arrow in space. It has a direction and a length called its <span class="term">magnitude</span>. A two-dimensional vector like [3, 2] means "go 3 units right and 2 units up". The arrow points somewhere specific. This view builds intuition.</p>
<p>The <span class="term">data view</span> treats a vector as an ordered list of measurements describing something: a person, a document, a job role, or a neural network input. This is how vectors live in AI and analytics every day.</p>
<p>Neither view is more correct than the other. The geometry helps me think. The data view connects to real work.</p>
</div>
<div class="inner-note">
<div class="inner-note-label">Professional Note</div>
<p>In real people analytics and machine learning work, vectors built from features on different scales are often standardised before magnitudes or distances are interpreted meaningfully. I do not need to master scaling yet, but I do need to avoid learning the wrong habit.</p>
</div>
<div class="inner-formal">
<div class="inner-formal-label">Formal Notation</div>
<p>A two-dimensional vector can be written as <strong style="color:#c4f042;">v = [v&#x2081;, v&#x2082;]</strong> or as a column vector. In NumPy it is typically stored as a one-dimensional array using <code class="inline">np.array([v1, v2])</code>.</p>
</div>
</div>

<div class="panel">
<div class="accent-strip accent-amber"></div>
<div class="sec-row">
<div class="sec-label amber">AI Connection</div>
<div class="sec-rule"></div>
</div>
<h3>How vectors power language models</h3>
<p>Every input to a neural network is a <span class="term">vector</span>. When a language model like GPT or Claude processes text, it does not work with words directly. It converts tokens into vectors called <span class="term">embeddings</span>, often hundreds of dimensions long. Tokens with similar meanings tend to sit closer together in that high-dimensional space, which is one reason the model can work with meaning numerically.</p>
<p>A recommendation system does something similar. It can represent a user's behaviour as a vector and compare that vector with other users or with item vectors to decide what to suggest next.</p>
</div>

<div class="panel">
<div class="accent-strip accent-amber"></div>
<div class="sec-row">
<div class="sec-label amber">Workforce Analytics Connection</div>
<div class="sec-rule"></div>
</div>
<h3>Representing people as points in space</h3>
<p>In people analytics, an employee can be represented as a <span class="term">feature vector</span> using attributes such as tenure, engagement, salary position, performance, or absence frequency. That vector becomes the input to an attrition model, a clustering workflow, or a similarity analysis. The model never sees a person directly. It sees a point in multi-dimensional space.</p>
<div class="inner-note">
<div class="inner-note-label">Professional Note</div>
<p>Those raw features sit on different scales, so magnitudes and distances only become meaningfully comparable once scaling is applied. Building that habit early matters.</p>
</div>
</div>

</div>

```python
import numpy             as np
import matplotlib.pyplot as plt

# ── Colour constants ───────────────────────────────────────────────────────────
BG    = '#0a0a0c'
TEXT  = '#e8e4de'
SUB   = '#b5b0a8'
MUTED = '#6a6560'
GRID  = '#2a2a28'
AXIS  = '#3a3a38'
GRN   = '#c4f042'
AMB   = '#f0a830'
RED   = '#e8453c'

# ── Scalars ───────────────────────────────────────────────────────────────────
age              = 34
engagement_score = 72

print(f"Age (scalar):              {age}")
print(f"Engagement score (scalar): {engagement_score}")

# ── Employee profile as a vector ──────────────────────────────────────────────
# Features: [age, tenure_years, engagement_score, performance_rating]
# These features sit on different scales, so raw magnitude is not analytically
# meaningful without standardisation. For now, I am just observing the structure.

employee_vector    = np.array([34, 5, 72, 4], dtype = float)
employee_magnitude = np.linalg.norm(employee_vector)

print(f"\nEmployee vector:    {employee_vector}")
print(f"Number of dimensions: {employee_vector.shape[0]}")
print(f"Raw magnitude:        {employee_magnitude:.4f}")
print(f"  (note: dominated by engagement_score at scale 0-100)")

# ── Geometric vector in 2D space ──────────────────────────────────────────────
v           = np.array([3, 2], dtype = float)
v_magnitude = np.linalg.norm(v)

print(f"\nGeometric vector v = {v}")
print(f"Magnitude |v|        = {v_magnitude:.4f}")

fig, ax = plt.subplots(figsize = (6, 6))
fig.patch.set_facecolor(BG)
ax.set_facecolor(BG)

ax.quiver(
    0, 0, v[0], v[1],
    angles      = 'xy',
    scale_units = 'xy',
    scale       = 1,
    color       = GRN,
    width       = 0.015
)

ax.set_xlim(-1, 5)
ax.set_ylim(-1, 5)
ax.set_aspect('equal', adjustable = 'box')

ax.axhline(0, color = AXIS, linewidth = 0.8)
ax.axvline(0, color = AXIS, linewidth = 0.8)
ax.grid(True, color = GRID, alpha = 0.4)

ax.set_title(
    "Vector v = [3, 2] as an arrow in 2D space",
    color    = TEXT,
    fontsize = 11,
    pad      = 14
)
ax.set_xlabel("x", color = SUB)
ax.set_ylabel("y", color = SUB)
ax.tick_params(colors = MUTED)

for spine in ax.spines.values():
    spine.set_edgecolor(AXIS)

ax.annotate(
    "v = [3, 2]",
    xy       = (3, 2),
    xytext   = (3.2, 1.7),
    color    = GRN,
    fontsize = 9
)

plt.tight_layout()
plt.show()
```
